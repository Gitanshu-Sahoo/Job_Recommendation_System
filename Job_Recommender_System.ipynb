```python
# ==========================================
# JOB RECOMMENDER SYSTEM
# ==========================================
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
user_skills = input("Enter your skills (separate with spaces):\n\n").lower()
all_skills = df["Skills"].tolist()
all_skills.append(user_skills)
vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform(all_skills)
similarity_matrix = cosine_similarity(vectors)
user_similarity = similarity_matrix[-1][:-1]
df["Match Score"] = user_similarity * 100
recommendations = df.sort_values(
    by="Match Score",
    ascending=False
)
# ==========================================
# MATCH SCORES
# ==========================================
print("\n" + "="*70)
print("SKILL MATCH SCORES")
print("="*70)
for index, row in recommendations.iterrows():
    print(
        f"{row['Job Title']} --> "
        f"{row['Match Score']:.2f}%"
    )
# ==========================================
# TOP RECOMMENDATIONS
# ==========================================
print("\n" + "="*70)
print("TOP 5 JOB RECOMMENDATIONS")
print("="*70)
for i, (_, row) in enumerate(
        recommendations.head(5).iterrows(),
        start=1):
    print(
        f"{i}. {row['Job Title']} "
        f"({row['Match Score']:.2f}% Match)"
    )
# ==========================================
# SKILL GAP ANALYSIS
# ==========================================
best_job = recommendations.iloc[0]
job_skills = set(
    best_job["Skills"].split()
)
user_skill_set = set(
    user_skills.split()
)
matched_skills = (
    user_skill_set.intersection(job_skills)
)
missing_skills = (
    job_skills - user_skill_set
)
print("\n" + "="*70)
print("SKILL GAP ANALYSIS")
print("="*70)
print("\nBest Matching Career:")
print(best_job["Job Title"])
print("\nSkills You Already Have:")
if len(matched_skills) > 0:
    for skill in matched_skills:
        print("✓", skill)
else:
    print("No matching skills found")
print("\nSkills To Learn:")
if len(missing_skills) > 0:
    for skill in missing_skills:
        print("✗", skill)
else:
    print("No skill gaps found!")
print("\n" + "="*70)
print("END OF REPORT")
print("="*70)
```
